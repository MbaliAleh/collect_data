#MAKING FUNCTIONS TO PICK DIFFERENT PARAMETERS IN data1 file

```python
#PICKING N-TIMES

n_l = npList_all

def func1_time():
    
   xt = []  #Building a list of time data series

   p=10
    
   for i in range(len(npList_all)):
       p = p +n_l[i] + 4
       #num = data1.loc[[ p], 'bmnum':'=']
       num = data1.iloc[p,1]  # Generate one random number
       xt.append(num)             # Save all these times in a list xt

   tot_t = xt
   return tot_t
```

```python
time_data = func1_time() #this returns a list of all the times as desired
time_data=pd.Series(time_data)
time_data.head()
```

```python
len(time_data)
```

```python
#PICKING BEAM NUMBERS

n_l=npList_all

def func_bm():
    
   xb = []
   p = 11  #Defining a starting point
    
   for i in range(len(n_l)):
    
       p = p + n_l[i] + 4 
       v = data1.iloc[p,2]
       xb.append(v) 
        
   tot_b = xb
   return tot_b
```

```python
beam_data = func_bm()
beam_data
```

```python
len(beam_data)
```

```python
#PICKING SKY NOISE LEVEL

n_l=npList_all

def func_skl():
    
   xs = []
   p = 11
    
    
   for i in range(len(n_l)):
       p = p + n_l[i] + 4 # Same as p = p+1
       v = data1.iloc[p,8]
       xs.append(v) 
        
   tot_s = xs
   return tot_s
```

```python
skyN_data = func_skl()
skyN_data
```

```python
len(skyN_data)
```

```python
#PICKING GATES

n_l=npList_all

n_k=npList1

def func_gates1():
   xg = []
   p = 14
   q = 23

   for i,j in zip(range(len(n_l)), range(len(n_k))):
       p = p + n_l[i] + 4 
       q = q + n_k[j] + 4
       v1 = data1.iloc[p:q,0].tolist()
       xg.append(v1)
   tot_g = xg
   return tot_g
```

```python
gate_data = func_gates1()
s1 = pd.Series(gate_data)
s1.head()
```

```python
#PICKING GATE RANGE

n_l=npList_all

n_k=npList1

def func_rng():
    
   xr = []

   p = 14
   q = 23

   for i,j in zip(range(len(n_l)), range(len(n_k))):
       p = p + n_l[i] + 4 
       q = q + n_k[j] + 4
       v1 = data1.iloc[p:q,1].tolist()
       xr.append(v1)
        
   tot_r = xr
   return tot_r
```
